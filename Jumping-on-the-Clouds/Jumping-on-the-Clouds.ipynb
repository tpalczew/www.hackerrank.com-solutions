```python
#!/bin/python3

import math
import os
import random
import re
import sys



def jumpingOnClouds(c):
    print(c)
    print(len(c))
    count = 0 
    ind = 0    
    while ind < len(c)-1:
        if c[ind+1]==1 and c[ind+2]==1:
            print('No Way')
            break
        if ind+2 < len(c) and c[ind+2]!=1:
            ind +=1
        ind +=1
        count+=1
    return count            
```

```python
n = int(input())

c = list(map(int, input().rstrip().split()))
```

```python
result = jumpingOnClouds(c)
```
