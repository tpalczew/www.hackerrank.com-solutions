```python
text = input()
listtext = list(text)


def solution():
    final = []
    count = 1
    for ind, elem in enumerate(text):
        if ind < len(text)-1:
            if listtext[ind+1] == elem:
                count+=1 
            else:
                count=count
                final.append( (count, int(elem)) )
                count = 1
        else:
            final.append( (count, int(elem)) )
    print( *final )

solution()
```
