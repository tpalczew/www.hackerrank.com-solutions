```python
def solution1():
    for _ in range(int(input())):
        
        ccard = input()
        
        notok = False
        
        # 16 digits
        dig = [n for n in ccard if n.isdigit()]
        digs = len( dig )
        if digs != 16:
            notok = True
        
        # only digits or - 
        symbl = [n for n in ccard if n.isdigit() or n=='-']
        for elem in ccard:
            if elem not in symbl:
                notok = True
        
        # starts with 4, or 5, or 6
        if int(ccard[:1]) not in [4,5,6]: 
            notok = True
        
        # equal groups of 4
        
        if "-" in ccard:
            try:
                cc1, cc2, cc3, cc4 = ccard.split('-')
                if len(cc1) != 4:
                    notok = True
                if len(cc2) != 4:
                    notok = True
                if len(cc3) != 4:
                    notok = True
                if len(cc4) != 4:
                    notok = True
            except:
                notok = True
        
        ccard = ccard.replace("-","")
        
        # NOT use any other separator like ' ' , '_', etc. 
        for i, n in enumerate(ccard):
            if (' ' in n) == True:
                notok = True
        
        # may have digits in groups of 4, separated by one hyphen "-"
        for i, n in enumerate(ccard):
            if i <= len(ccard)-4:
                if (ccard[i], ccard[i+1], ccard[i+2], ccard[i+3]) == (n, n, n, n):
                    notok = True
        
        if notok:
            print("Invalid")
        else:
            print("Valid")
```

```python
def solution2():
    import re
    PATTERN = "([4-6]{1})([0-9]{3}-{0,1})([0-9]{4}-{0,1})([0-9]{4}-{0,1})([0-9]{4})"
    for _ in range(int(input())):
        ccard = input()
        if bool(re.match(PATTERN, ccard)) is False:
            print("Invalid")
            continue
        ccard = ccard.replace("-","")
        notok = False
        for i, n in enumerate(ccard):
            if (' ' in n) == True:
                notok = True
        for i, n in enumerate(ccard):
            if i <= len(ccard)-4:
                if (ccard[i], ccard[i+1], ccard[i+2], ccard[i+3]) == (n, n, n, n):
                    notok = True
        if notok:
            print("Invalid")
        else:
            print("Valid")
```
