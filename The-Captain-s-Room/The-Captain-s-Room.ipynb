```python
n = int(input())
numbers = (input().split(" "))

def solution(numbers):
    numbers = list(numbers)
    setnumbers = set(numbers)
    for elem in setnumbers:
        try:
            numbers.remove(elem)
        except:
            pass
    print(  *(setnumbers - set(numbers) ) )

solution(numbers)
```

```python
n = int(input())
numbers = (input().split(" "))

def solution2(numbers): 
    numbers = [int(i) for i in numbers]   
    myset = set(map(int,numbers))
    myset = [n*i for i in myset]
    fin = ( sum(myset) - sum(numbers)) / (n-1)  
    print(int(fin))

solution2(numbers)
```
