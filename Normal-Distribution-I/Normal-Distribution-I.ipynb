```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

# The first line contains 2 space-separated values denoting the respective mean and 
# standard deviation for X. 
# The second line contains the number associated with question 1. 
# The third line contains 2 space-separated values describing the 
# respective lower and upper range boundaries for question 2.

import math

mean, sigma = map(int, input().split())
q1 = float(input())
q2_lower, q2_upper = map(float, input().split())

debug = 0

if debug == 1:
    print('mean = ' + str(mean) )
    print('sigma = ' + str(sigma))
    print('q1 [0,' + str(q1) + "]")
    print('q2 = [' + str(q2_lower) + ',' + str(q2_upper) + ']')


def phi(miu, sigma, x):
    'Cumulative distribution function for a function with normal distribution'
    return (1/2) * ( 1.0 + math.erf( (x - miu) / ( sigma * math.sqrt(2.0) ) ) )

q1answer = phi(mean, sigma, q1)
q1answer = round(q1answer, 3)

q2answer = phi(mean, sigma, q2_upper) - phi(mean, sigma, q2_lower)
q2answer = round(q2answer, 3)
print(q1answer)
print(q2answer)
```
