```python
import math

size_of_datasets = int(input())

X = list(map(float, input().split()))
Y = list(map(float, input().split()))

Xmean = sum(X)/size_of_datasets
Ymean = sum(Y)/size_of_datasets

Xstd_part = [(i - Xmean)**2 for i in X]
Xstd = math.sqrt(sum(Xstd_part)/size_of_datasets)

Ystd_part = [(i - Ymean)**2 for i in Y]
Ystd = math.sqrt(sum(Ystd_part)/size_of_datasets)

PearsonCorrelationCoefficient = sum([(i-Xmean)*(j-Ymean) for i,j in zip(X,Y)])/ (size_of_datasets*Xstd*Ystd)

print(round(PearsonCorrelationCoefficient,3))
```
