```python
import email.utils
import re
pattern = r"^[A-Za-z](\w|-|\.|_)+@[A-Za-z]+\.[A-Za-z]{1,3}$"

m = int(input())
for el in range(m):
    checkemail = email.utils.parseaddr(input())
    if re.match(pattern, checkemail[1]):
        print(email.utils.formataddr(checkemail))
```
