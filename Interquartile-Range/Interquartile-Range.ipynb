```python
'''
The first line contains an integer n denoting the number of elements in arrays X and F. 
The second line contains n space-separated integers describing the respective elements of array X. 
The third line contains n space-separated integers describing the respective elements of array F. F represents the respective frequencie.
'''
import math
n = int(input())
X = list(map(int, input().split() ))
F = list(map(int, input().split() ))
S = []
for ind, elem in enumerate(F):
    for i in range(elem):
        S.append(X[ind])
S = sorted(S)
l = math.floor(len(S)/2)
r = math.floor((len(S)+1)/2)
S_l = S[:l]
S_r = S[r:]

def med(arr, m):
    if m%2 == 0:
        l = int(m/2)-1
        r = int(m/2)
        return (arr[l] + arr[r])/2
    else:
        v = math.floor(m/2)
        return (arr[v])

medl = med(S_l, len(S_l))
medr = med(S_r, len(S_r))

print("{:.1f}".format(medr - medl))
```
