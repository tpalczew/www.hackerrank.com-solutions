```python
def merge_the_tools(string, k):
    # your code goes here
    lenstring = len(string)
    print("len string (n) = " + str(lenstring))
    print("k = " + str(k))
    split = int(lenstring/k)
    print("n/k = " + str(split))
    chunk_size = int(lenstring/split)
    print("chunk size = " + str(chunk_size))
    splitted_string = [string[i:i+chunk_size] for i in range(0, len(string), chunk_size)]
    
    # PRESERVING ORDER IN PYTHON
    def unique(sequence):
        seen = set()
        return [x for x in sequence if not (x in seen or seen.add(x))]
    
    for element in splitted_string:
        element = unique(element)
        elem = ''.join(element)
        print(elem)
```

```python
string = "AABCAAADA"
k = 3 
```

```python
merge_the_tools(string,k)
```
