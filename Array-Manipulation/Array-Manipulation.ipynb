```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the arrayManipulation function below.
def arrayManipulation2(n, queries):  
    # ok you can solve it but now let's go back and try in place or with less memory 
    empty = [0 for i in range(n)] 
    for elem in queries:
        for x in range(elem[0], elem[1]+1):
            empty[x-1] = empty[x-1] + elem[2]
    return max(empty)

def arrayManipulation(n, queries):  
    from itertools import accumulate
    dx = [0] * (n + 1)
    for elem in queries:
        dx[elem[0] - 1] += elem[2]
        dx[elem[1]] -= elem[2]
    return max(accumulate(dx))
```

```python
nm = input().split()

n = int(nm[0])

m = int(nm[1])

queries = []

for _ in range(m):
    queries.append(list(map(int, input().rstrip().split())))
```
