```python
from collections import deque
```

```python
def solution():
    n = int(input())
    for elem in range(n):
        d = deque()
        n = int(input())
        values = list(map(int, input().split()))
        d.extend(values)
    
        for v in reversed(sorted(d)):
            # we want to go from the largest to the smallest values and check edges
            if d[-1] == v: # right
                d.pop()
            elif d[0] == v: # left
                d.popleft()
            else:
                print('No')
                break
        else: 
            print('Yes')
```
