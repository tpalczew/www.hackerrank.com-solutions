```python
import math
import os
import random
import re
import sys

# Complete the beautifulDays function below.
def beautifulDays(i, j, k):
    count_beautiful = 0
    for elem in range(i,j+1):
        # inclusive range
        elemrev = str(elem)     # make string
        elemrev = elemrev[::-1] # reverse
        elemrev = int(elemrev)  # make int
        if abs(elem - elemrev) % k == 0:
             count_beautiful += 1
    return count_beautiful
```

```python
ijk = input().split()

i = int(ijk[0])

j = int(ijk[1])

k = int(ijk[2])

result = beautifulDays(i, j, k)
```
