```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import Counter

os.environ["OUTPUT_PATH"] = "/Users/tpalczew/test.txt"


# Complete the birthdayCakeCandles function below.
def birthdayCakeCandles(ar):
    countcandles = Counter(ar)
    tallest = countcandles.most_common(1) 
    for value, count in tallest: final = count
    return final

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input())

    ar = list(map(int, input().rstrip().split()))

    result = birthdayCakeCandles(ar)
    
    print(result)

    fptr.write(str(result) + '\n')

    fptr.close()
```
