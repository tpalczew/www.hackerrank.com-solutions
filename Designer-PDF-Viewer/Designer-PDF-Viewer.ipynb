```python
#!/bin/python3

import math
import os
import random
import re
import sys

from collections import OrderedDict
import string

# Complete the designerPdfViewer function below.
def designerPdfViewer2(h, word):
    ordword = OrderedDict(zip(string.ascii_lowercase, h)) #you don't really need ordered dictionary 
    #ordword = dict(zip(string.ascii_lowercase, h))
    print(ordword)
    t = []
    for elem in word:
        t.append(ordword[elem])
    return (max(t)*len(word))

def designerPdfViewer(h, word):
    # if you don't want to use string
    letters = 'abcdefghijklmnopqrstuvwxyz'
    ordword = dict(zip(letters, h))
    print(ordword)
    # list comprehension instead of for loop 
    return (len(word) * max( [ordword[x] for x in ordword if x in word] ))
```
