```python
#!/bin/python3

import math
import os
import random
import re
import sys



#if __name__ == '__main__':

def solution():
    nm = input().split()

    n = int(nm[0])
    m = int(nm[1])

    matrix = []

    for _ in range(n):
        matrix_item = input()
        matrix.append(matrix_item)
    
    fmatrix = []
    for ms in range(m):
        for ns in range(n):
            fmatrix.append(matrix[ns][ms:ms+1])
                        
    fmatrix = "".join(fmatrix)       
    
    fmatrix = re.sub(r'(?<=[a-zA-Z0-9])[^a-zA-Z0-9]+(?=[a-zA-Z0-9])',' ',fmatrix)
    
    print(fmatrix)
    
solution()
```
