```python
from collections import namedtuple

def solution():
    
    n = int(input())
    classes = input().split()
    
    try:
        TestCase = namedtuple('TestCase', classes)
    except ValueError as e:
        print(e)
        pass
        
    total = 0
    for element in range(n):
        try:
            field1, field2, field3, field4 = input().split()
            student = TestCase(field1,field2,field3,field4)
        except: 
            print("Different number of classes (not 4) in the testcase, something may not work, check your testcase")
            break
        try:
            total += int(student.MARKS)
            print('{:.2f}'.format(total/n))
        except:
            print("MARKS info is probably missing, check your testcase")
            break
    
```
