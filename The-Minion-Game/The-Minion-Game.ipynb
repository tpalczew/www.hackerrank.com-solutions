```python
def minion_game(string):
    # your code goes here
    
    stuart, kevin = 0, 0
    vowels = ("A", "E", "I", "O", "U")    
    
    for i in range(len(string)):
        if string[i] in vowels:
            kevin += (len(string)-i)
        else:
            stuart += (len(string)-i)
        
    if kevin > stuart:
        print("Kevin " + str(kevin))
    if stuart > kevin:
        print("Stuart " + str(stuart))
    if kevin == stuart:
        print("Draw")
```

```python
minion_game('BANANA')
```
