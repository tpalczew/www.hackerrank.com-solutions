```python
#!/bin/python3

import math
import os
import random
import re
import sys


os.environ["OUTPUT_PATH"] = "/Users/tpalczew/test.txt"


# Complete the minimumSwaps function below.
def minimumSwaps(arr):
    last_index = len(arr)-1
    index_dict = {v: i for i,v in enumerate(arr)} # use dictionary to lookup index for specific value 
    swaps = 0
    for i, v in enumerate(arr):
        if v != i+1: # if value is different than it should be (i.e value 1 -> index 0 + 1 = 1)
            to_swap = index_dict[i+1] # check what to swap by value (from dict)
            arr[to_swap] = arr[i]
            arr[i] = arr[to_swap]
            index_dict[v] = to_swap
            index_dict[i+1] = i
            swaps += 1
    return swaps
            


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    res = minimumSwaps(arr)
    
    print(res)

    fptr.write(str(res) + '\n')

    fptr.close()
```
