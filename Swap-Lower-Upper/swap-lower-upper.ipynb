```python
def swap_case(s):
    new_s = []
    for element in s:
        if element.isupper():
            element = element.lower()
        else:
            element = element.upper()
        new_s.append(element)
    return str(''.join(new_s))
```

```python
if __name__ == '__main__':
    s = input()
    
    if isinstance(s, str):
        result = swap_case(s)
        print(result)
    else:
        print('We need a string')
```
