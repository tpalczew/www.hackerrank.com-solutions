```python
def solution():
    for _ in range(int(input())):
        uid = input()
        
        import re
        upper_case_l = len([i for i in uid if i.isupper()])
        digits_n =  len([i for i in uid if i.isdigit()])
        
        if len(uid) != 10:
            print("Invalid")
            continue
        if len(set(uid)) != 10:
            print("Invalid")
            continue
        if not uid.isalnum():
            print("Invalid")
            continue
        if upper_case_l < 2:
            print("Invalid")
            continue
        if digits_n < 3:
            print("Invalid")
            continue
    
        print("Valid")
solution()
```
