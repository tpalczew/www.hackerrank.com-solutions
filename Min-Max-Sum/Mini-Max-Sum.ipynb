```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import deque

# Complete the miniMaxSum function below.
def miniMaxSum(arr):
    de1 = deque(arr)
    de2 = deque(arr)
    minv = 0
    maxv = 0 
    for i in range(len(arr)-1):
        minv = minv + min(de1)
        maxv = maxv + max(de2)
        de1.remove(min(de1))
        de2.remove(max(de2))
    print(minv, maxv)

if __name__ == '__main__':
    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```
