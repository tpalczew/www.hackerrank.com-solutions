```python
#!/bin/python3

import math
import os
import random
import re
import sys
```

```python
def solve(s):
    
    s = list(s)
    if s[0].isdigit():
        s[0] = s[0]
    else:
        s[0] = s[0].upper()
    
    indtochange = []
    doit = False
    
    for ind, elem in enumerate(s):
        if elem == ' ':
            indtochange.append(ind+1)
            doit = True
            
    if doit:
        for inde in indtochange:
            s[inde] = s[inde].upper()
    
    s = "".join(s)
    return s
```

```python
solve("tom jan palczewski")
```
