```python
#!/bin/python3

import math
import os
import random
import re
import sys
import time
import datetime
from dateutil.tz import gettz, tzoffset


class time_class:

    # Initializer / Instance Attribute
    def __init__(self, day, month, year, hour, minute, second, zone1, zone2):
        self.day = day
        self.month = month
        self.year = year
        
        self.hour = hour
        self.minute = minute
        self.second = second
        
        self.zone1 = zone1
        self.zone2 = zone2
        
    # Instance Method
    def time_difference(self, other):
        # Sun 10 May 2015 13:54:36 -0700
        # tzinfo=datetime.timezone(datetime.timedelta(-1, 72000))
        # tzinfo=tzoffset(None, 3600)
        
        t1 = datetime.datetime(int(self.year), int(self.month), int(self.day), int(self.hour), int(self.minute), int(self.second), tzinfo=tzoffset(None, int(self.zone1)*int(self.zone2)))
        
        
        t2 = datetime.datetime(int(other.year), int(other.month), int(other.day), int(other.hour), int(other.minute), int(other.second), tzinfo=tzoffset(None, int(other.zone1)*int(other.zone2)))
        
        
        return str(int(abs((t1-t2).total_seconds())))
        
# Complete the time_delta function below.
def time_delta(t1, t2):
    
    datetimeflag = False
    oldschool = True
    
    if oldschool == True:
        days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
        months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        
        t1l = t1.split()
        
        # Sun 10 May 2015 13:54:36 -0700
        t1d = t1l[0]
        t1dd = t1l[1]
        t1mm = t1l[2]
        t1y = t1l[3]
        
        t1t = t1l[4]
        t1h, t1m, t1s = t1t.split(":")
        
        t1mm = months.index(t1mm) + 1 # month number 1-12
        
        t1z = t1l[5]
        t1zone_sign = int( t1z[:1] + '1' ) # time zone difference sign
        t1zonehours = int(t1z[1:3])
        t1zoneminutes = int(t1z[3:])
        t1z2 = (t1zonehours * 60 * 60) + (t1zoneminutes * 60) # time zone difference in seconds
        
        
        time1 = time_class(t1dd, t1mm, t1y, t1h, t1m, t1s, t1zone_sign, t1z2)
        
        t2l = t2.split()
        
        t2d = t2l[0]
        t2dd = t2l[1]
        t2mm = t2l[2]
        t2y = t2l[3]
        t2t = t2l[4]
        t2h, t2m, t2s = t2t.split(":")
        t2mm = months.index(t2mm) + 1 # month number 1-12
        t2z = t2l[5]
        t2zone_sign = int( t2z[:1] + '1' ) # time zone difference sign
        t2zonehours = int(t2z[1:3])
        t2zoneminutes = int(t2z[3:])
        t2z2 = (t2zonehours * 60 * 60) + (t2zoneminutes * 60) # time zone difference in seconds
        
        time2 = time_class(t2dd, t2mm, t2y, t2h, t2m, t2s, t2zone_sign, t2z2)
        
        # use Instance Method to get time difference in s
        delta_t = time1.time_difference(time2)
        
        return str(delta_t)
        
                
    if datetimeflag == True:
        t1=datetime.datetime.strptime(t1,'%a %d %b %Y %H:%M:%S %z')
        t2=datetime.datetime.strptime(t2,'%a %d %b %Y %H:%M:%S %z')
        return str(int(abs((t1-t2).total_seconds())))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()
        t2 = input()
        delta = time_delta(t1, t2)
        print(delta)
        fptr.write(delta + '\n')
    fptr.close()
```
