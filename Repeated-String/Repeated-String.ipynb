```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import Counter

# Complete the repeatedString function below.
def repeatedString(s, n):
    if n > len(s):
        counter = 0 
        counts = Counter(s)
        numsa = counts['a']
        #
        if n%len(s)==0:
            counter = (n/len(s)) * numsa
        else:
            counter = (n//len(s)) * numsa
            leng = n - (n//len(s))*len(s) 
            for elem in range(leng):
                if s[elem] == 'a':
                    counter = counter + 1
    else:
        counts = Counter(s[:n])
        counter = counts['a']              
    return int(counter)
```

```python
s = input()

n = int(input())
```

```python
result = repeatedString(s, n)
```
