```python
import itertools
st, k = (input().split(" "))
st = list(st)
st = sorted(st)
k = int(k)


for elem in range(k):
    elem = elem + 1 
    perm = itertools.combinations(st, elem)
    for elem in perm:
        elem = ''.join(map(str,elem))
        print(elem)
```
