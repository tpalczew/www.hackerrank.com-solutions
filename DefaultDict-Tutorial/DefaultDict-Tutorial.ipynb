```python
from collections import defaultdict
```

```python
def solution():
    n, m = input().split(" ")
    d = defaultdict(list)
    indx = 1
    for en in range(int(n)):
        d[input()].append(indx)
        indx+=1
    for en in range(int(m)):
        check = input()
        if check not in d:
            print(-1)
        if check in d:
            print(*d[check]) 

solution()
```
