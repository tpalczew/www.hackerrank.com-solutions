```python
from collections import deque
d = deque()

n = int(input())

for elem in range(n):
    com, *e = input().split()
    if e:
        doit = "d." + str(com) + "(" + str(e) + ")" 
    else:
        doit = "d." + str(com) + "()"
    exec(doit)

print(*[" ".join(e) for e in d] )
```
