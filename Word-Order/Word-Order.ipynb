```python
from collections import OrderedDict

m = int( input() )
#words = [ input() for elem in range(m) ]
ordered_dict = OrderedDict()
for elem in range(m):
    word = input()
    ordered_dict[word] = ordered_dict.get(word, 0) + 1

print(len(ordered_dict))
print(*ordered_dict.values())
```
